```python
# question 2 data cleaning

data = all_parties[endorsement_columns + ['Primary %']]
data.loc[:, endorsement_columns]= data[endorsement_columns].fillna('No').replace({'No': 0, 'Yes': 1})
for col in endorsement_columns:
    data[col] = pd.to_numeric(data[col])
data = data.dropna(subset=['Primary %'])
data.head(5)
```

```python
# question 2 Train-Test Split
from sklearn.model_selection import train_test_split

train, test = train_test_split(data, test_size=0.3, random_state=101)
```

```python
# question 2 GLM Model

model = sm.GLM(train['Primary %'], exog=sm.add_constant(train[endorsement_columns]), family=sm.families.Gaussian())
results = model.fit()

train["glm_pred"] = results.predict(sm.add_constant(train[endorsement_columns]))
test["glm_pred"] = results.predict(sm.add_constant(test[endorsement_columns]))

print(results.summary())
```

```python
# question 2 GLM Model Metrics
glm_train_rmse = np.mean((train["glm_pred"] - train["Primary %"]) ** 2) ** 0.5
glm_test_rmse = np.mean((test["glm_pred"] - test["Primary %"]) ** 2) ** 0.5

glm_train_r2 = metrics.r2_score(train['Primary %'], train["glm_pred"])
glm_test_r2 = metrics.r2_score(test['Primary %'], test["glm_pred"])

print(f"GLM Training RMSE: {glm_train_rmse:.4f}")
print(f"GLM Test RMSE:     {glm_test_rmse:.4f}")
print(f"GLM Training R²:   {glm_train_r2:.4f}")
print(f"GLM Test R²:       {glm_test_r2:.4f}")
print(f"GLM AIC:           {results.aic:.4f}")
```

```python
# question 2 Decision Tree Model
from sklearn.tree import DecisionTreeRegressor

tree_model = DecisionTreeRegressor()
tree_model.fit(train[endorsement_columns], train['Primary %'])

train["tree_pred"] = tree_model.predict(train[endorsement_columns])
test["tree_pred"] = tree_model.predict(test[endorsement_columns])
```

```python
# question 2 Decision Tree Plot
from sklearn.tree import plot_tree

plt.figure(figsize=(16, 4))
plot_tree(tree_model);
```

```python
# question 2 Decision Tree Metrics
tree_train_rmse = np.mean((train["tree_pred"] - train["Primary %"]) ** 2) ** 0.5
tree_test_rmse = np.mean((test["tree_pred"] - test["Primary %"]) ** 2) ** 0.5

tree_train_r2 = metrics.r2_score(train['Primary %'], train['tree_pred'])
tree_test_r2 = metrics.r2_score(test['Primary %'], test['tree_pred'])

tree_rss = np.sum((train["Primary %"] - train['tree_pred']) ** 2)
tree_aic = 2 * tree_model.get_n_leaves() + len(train) * np.log(tree_rss / len(train))

print(f"Decision Tree Training RMSE: {tree_train_rmse:.4f}")
print(f"Decision Tree Test RMSE:     {tree_test_rmse:.4f}")
print(f"Decision Tree Training R²:   {tree_train_r2:.4f}")
print(f"Decision Tree Test R²:       {tree_test_r2:.4f}")
print(f"Decision Tree AIC:           {tree_aic:.4f}")
```

#### Methods 

- We are aiming to predict the primary vote percentage by using endorsement type and count as features. We selected these features because endorsement types directly affect a voter’s perception of the candidate and the candidate’s funding, thus the candidate’s vote percentage, and the count affects the magnitude of these endorsements.

- For the GLM, we are using a linear regression model because we are trying to predict a real-valued output (primary %). We assume a linear relationship between the predictors and target variable, and each data entry is independent of one another.

- For the non-parametric method, we are using a decision tree because we wanted a simple model with high interpretability, especially since we are working with a small dataset. Also, we made this selection because of its ability to handle categorical features in prediction. We assume that our training data is representative of our problem and that our data is unbiased, or else our model will produce a poor estimate on test data.

- In evaluating the performance of the GLM and the decision tree, we will be using RMSE, R², and AIC.

#### Results

- The GLM achieved training and testing RMSE values of 28.7748 and 27.6585, respectively; training and testing R² values of 0.0610 and 0.0622, respectively; and an AIC value of 10564.1254. The RMSE and R² values are very close to each other suggesting it is not overfitting; however, the high AIC score suggests a suboptimal model for the data, being too simple.

- Low R^2 value for the GLM (0.06) suggest that predictor (endorsement type and count) may not be sufficient to explain the target variable (primary vote percentage) well 

    - Even if coefficients are statistically significant, their practical utility on explaining the primary vote percentage is limited 

- The Decision Tree achieved training and testing RMSE values of 28.4070 and 28.1972, respectively; training and testing R² values of 0.0849 and 0.0253, respectively; and an AIC value of 7497.2893. The RMSE values are very close to each other; however, the R² values are different from each other, suggesting the model is overfitting, and does not generalize well to unseen data. Additionally, the low AIC score suggests a better fit model, fitting better to the training data.

- GLM Uncertainty: Because we are doing a frequentist approach, our coefficient estimates are based on the observed sample data and may vary if we were to collect new samples due to the inherent randomness in the data. The generated confidence interval for each coefficient has a 95% chance of containing our true coefficient. However, with our low performance on our model, these coefficients may not be as helpful in explaining our target variable.

#### Discussion

- On the training data, the Decision Tree performed better because it had a lower RMSE value and higher R² value when compared to the GLM, indicating that the model made more accurate predictions and explained more variability in the data. However, on the testing data, the GLM performed better because it had a lower RMSE value and higher R² value when compared to the Decision Tree, indicating that on unseen data the model made more accurate predictions and explained more variability in the data. Additionally, the Decision Tree received a higher AIC value than the GLM, reflecting that the GLM is a simpler model than the Decision Tree, but showing its simplicity helps in preventing overfitting.
